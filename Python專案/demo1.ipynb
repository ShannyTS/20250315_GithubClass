- 這裡是一個建立`正弦波`圖表的範例


```python
import numpy as np
import matplotlib.pyplot as plt

# 設定 x 軸的範圍和取樣點數量
x = np.linspace(0, 2 * np.pi, 100)

# 計算 y 值，即 sin(x)
y = np.sin(x)

# 使用 matplotlib 繪製圖表
plt.plot(x, y)
plt.title('Sin plot')  # 設定圖表標題
plt.xlabel('x')  # 設定 x 軸標籤
plt.ylabel('sin(x)')  # 設定 y 軸標籤
plt.grid(True)  # 顯示格線
plt.show()  # 顯示圖表
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 設定 x 軸的範圍和取樣點數量
x = np.linspace(0, 2 * np.pi, 100)

# 計算 y 值，即 cos(x)
y = np.cos(x)

# 使用 matplotlib 繪製圖表
plt.plot(x, y)
plt.title('Cos plot')  # 設定圖表標題
plt.xlabel('x')  # 設定 x 軸標籤
plt.ylabel('cos(x)')  # 設定 y 軸標籤
plt.grid(True)  # 顯示格線
plt.show()  # 顯示圖表
```
