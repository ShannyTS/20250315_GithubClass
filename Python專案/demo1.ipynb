- 這裡是一個建立`正弦波`圖表的範例


```python
import numpy as np
import matplotlib.pyplot as plt

# 設定 x 軸的範圍和取樣點數量
x = np.linspace(0, 2 * np.pi, 100)

# 計算 y 值，即 sin(x)
y = np.sin(x)

# 使用 matplotlib 繪製圖表
plt.plot(x, y)
plt.title('Sin plot')  # 設定圖表標題
plt.xlabel('x')  # 設定 x 軸標籤
plt.ylabel('sin(x)')  # 設定 y 軸標籤
plt.grid(True)  # 顯示格線
plt.show()  # 顯示圖表
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 設定 x 軸的範圍和取樣點數量
x = np.linspace(0, 2 * np.pi, 100)

# 計算 y 值，即 cos(x)
y = np.cos(x)

# 使用 matplotlib 繪製圖表
plt.plot(x, y)
plt.title('Cos plot')  # 設定圖表標題
plt.xlabel('x')  # 設定 x 軸標籤
plt.ylabel('cos(x)')  # 設定 y 軸標籤
plt.grid(True)  # 顯示格線
plt.show()  # 顯示圖表
```

正切圖表

```python
import numpy as np
import matplotlib.pyplot as plt

# 設定 x 軸的範圍和取樣點數量，避開正切函數的漸近線
x = np.linspace(-np.pi/2 + 0.1, np.pi/2 - 0.1, 100)

# 計算 y 值，即 tan(x)
y = np.tan(x)

# 使用 matplotlib 繪製圖表
plt.plot(x, y)
plt.title('Tan plot')  # 設定圖表標題
plt.xlabel('x')  # 設定 x 軸標籤
plt.ylabel('tan(x)')  # 設定 y 軸標籤
plt.grid(True)  # 顯示格線
plt.show()  # 顯示圖表
```

-同時顯示 sin & cos的正弦波和餘弦波的圖表

```python
import numpy as np
import matplotlib.pyplot as plt

# 生成數據
x = np.linspace(0, 2 * np.pi, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)

# 繪製正弦波和餘弦波
plt.figure(figsize=(10, 5))
plt.plot(x, y_sin, label='sin(x)')
plt.plot(x, y_cos, label='cos(x)')
plt.title('sin & cos')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
